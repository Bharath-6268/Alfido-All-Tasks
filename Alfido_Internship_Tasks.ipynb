```python
# TASK-1

print("===== BASIC CALCULATOR =====")

while True:
    try:
        num1 = float(input("\nEnter first number: "))
        num2 = float(input("Enter second number: "))
    except ValueError:
        print("‚ùå Invalid input! Please enter valid numbers.")
        continue

    print("\nSelect Operation:")
    print("1. Addition (+)")
    print("2. Subtraction (-)")
    print("3. Multiplication (*)")
    print("4. Division (/)")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == "1":
        print(" Result:", num1 + num2)

    elif choice == "2":
        print(" Result:", num1 - num2)

    elif choice == "3":
        print(" Result:", num1 * num2)

    elif choice == "4":
        if num2 == 0:
            print(" Error: Division by zero is not allowed.")
        else:
            print(" Result:", num1 / num2)

    elif choice == "5":
        print(" Calculator closed successfully.")
        break

    else:
        print(" Invalid operation choice. Please select between 1 to 5.")
```

```python
#TASK-2

import random

print("===== NUMBER GUESSING GAME =====")
print("Guess a number between 1 and 100")

while True:
    secret_number = random.randint(1, 100)
    attempts = 0

    while True:
        try:
            guess = int(input("\nEnter your guess: "))
            attempts += 1
        except ValueError:
            print("Invalid input. Please enter a number only.")
            continue

        if guess < 1 or guess > 100:
            print("Number must be between 1 and 100.")
        elif guess < secret_number:
            print("Too low. Try again.")
        elif guess > secret_number:
            print("Too high. Try again.")
        else:
            print("You guessed the number correctly.")
            print("Total attempts:", attempts)
            break

    play_again = input("\nDo you want to play again? (yes/no): ").lower()
    if play_again != "yes":
        print("Thank you for playing.")
        break
```

```python
#TASK - 3
print("===== TEMPERATURE CONVERTER =====")

while True:
    print("\nSelect Conversion Type:")
    print("1. Celsius to Fahrenheit and Kelvin")
    print("2. Fahrenheit to Celsius and Kelvin")
    print("3. Kelvin to Celsius and Fahrenheit")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == "4":
        print("Program exited successfully.")
        break

    try:
        temp = float(input("Enter the temperature value: "))
    except ValueError:
        print("Invalid input. Please enter a valid number.")
        continue

    if choice == "1":
        fahrenheit = (temp * 9/5) + 32
        kelvin = temp + 273.15
        print("Fahrenheit:", fahrenheit)
        print("Kelvin:", kelvin)

    elif choice == "2":
        celsius = (temp - 32) * 5/9
        kelvin = celsius + 273.15
        print("Celsius:", celsius)
        print("Kelvin:", kelvin)

    elif choice == "3":
        if temp < 0:
            print("Kelvin cannot be negative.")
        else:
            celsius = temp - 273.15
            fahrenheit = (celsius * 9/5) + 32
            print("Celsius:", celsius)
            print("Fahrenheit:", fahrenheit)

    else:
        print("Invalid choice. Please select between 1 to 4.")
```

```python
# TASK - 4

print("===== SIMPLE TO-DO LIST APP =====")

tasks = []

while True:
    print("\nSelect an option:")
    print("1. Add a task")
    print("2. Remove a task")
    print("3. View all tasks")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == "1":
        task = input("Enter the task to add: ").strip()
        if task:
            tasks.append(task)
            print("Task added successfully.")
        else:
            print("Task cannot be empty.")

    elif choice == "2":
        if not tasks:
            print("No tasks to remove.")
            continue

        print("\nCurrent Tasks:")
        for i, t in enumerate(tasks, 1):
            print(f"{i}. {t}")

        try:
            index = int(input("Enter the task number to remove: "))
            if 1 <= index <= len(tasks):
                removed = tasks.pop(index - 1)
                print(f"Removed task: {removed}")
            else:
                print("Invalid task number.")
        except ValueError:
            print("Please enter a valid number.")

    elif choice == "3":
        if tasks:
            print("\nYour Tasks:")
            for i, t in enumerate(tasks, 1):
                print(f"{i}. {t}")
        else:
            print("No tasks available.")

    elif choice == "4":
        print("Exiting the To-Do List App.")
        break

    else:
        print("Invalid choice. Please select an option between 1 and 4.")
```
